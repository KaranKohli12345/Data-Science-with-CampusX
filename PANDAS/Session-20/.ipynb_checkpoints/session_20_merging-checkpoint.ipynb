## Joins

### inner join

```python
students
```

```python
registrations
```

```python
students.merge(registrations, how='inner', on='student_id')
```

### left join

```python
courses
```

```python
courses.merge(registrations, how='left', on='course_id')
```

```python
courses.merge(registrations, how='inner', on='course_id')
```

### right join

```python
temp_df = pd.DataFrame({
    'student_id': [26,27,28],
    'name': ['Karan', 'Nitish', 'Rahul'],
    'partner': [28,26,17]
})

temp_df
```

```python
students = pd.concat((students, temp_df), ignore_index=True)
students
```

```python
students.merge(registrations, how='right', on='student_id')
```

```python
students.merge(registrations, how='left', on='student_id')
```

### outer join

```python
students.merge(registrations, how='outer', on='student_id').tail(10)
```

- 99% inner join is used
- default is inner join if we don't mention it
