## Plotting 3D graphs / Meshgrids

- used to plot 3d graphs

```python
# Meshgrids are used to plot coordinates which are further feeded into a function to generate 3D plot.
```

```python
a = np.linspace(-10,9,20)
b = np.linspace(-10,9,20)
```

```python
a,b
```

```python
xx,yy = np.meshgrid(a,b)
```

```python
xx,yy
```

```python
import matplotlib.pyplot as plt
plt.scatter(xx,yy)
```

```python
def func(x,y):
    # return x**2 + y**2
    # return 2*x + 3*y
    # return 2*np.sin(x) + 3*y
    # return 2*np.log(x) + 3*y
    return 2*np.log(x) + 3*y**2

zz = func(xx,yy)
zz
```

```python
import plotly.express as px
import plotly.graph_objects as go

fig = px.scatter_3d()

fig.add_trace(go.Surface(x=xx,y=yy,z=zz))

fig.show()
```

```python
# name, iq, cgpa, placed

dt = np.dtype(
    [
        ('name', 'U20'),
        ('iq', np.int32),
        ('cgpa', np.float64),
        ('placed', 'U20'),
    ]
)
```

```python
dt
```

```python
students = np.array(
    [
        ('A', 100, 6.5, 'Yes'),
        ('B', 120, 9.7, 'Yes'),
        ('C', 90, 6.7, 'No')
    ], dtype=dt
)

students
```

```python
type(students[0][0])
```

```python
students[1][1]/10
```

```python
students['name']
```

```python
students['iq']
```
