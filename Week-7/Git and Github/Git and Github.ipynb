### Making changes

- git add 'file name' : adds file to the staging index
- git commit -m "message passed while commiting" : commits (add files to repo)

### when to commit?

- based on intuition
- When we have developed an independent feature of the software 
- for example a navbar of a website
- do it once for every independent feature
- do not do commit for multiple indepedent features
- for example a navbar and a sidebar of a website

### commit messages?

- short
- explain what is achieved
- dont use and word because and is used to add multiple featues which is not allowed
- syntax: This commit will ...
- remove first three words
- for example: This commit will add login functionality to the website.
- so the commit message will be "add login functionality to the website."
- dont explain why part only what part
