### plotting a simple function

```python
price = [45000, 54000, 34000, 49000, 60000, 55000]
year = [2015, 2016, 2017, 2018, 2019, 2020]

plt.plot(year, price)
```

```python
# from a pandas dataframe

batsman = pd.read_csv(r'datasets/sharma-kohli.csv')
batsman
```

```python
plt.plot(batsman['index'], batsman['V Kohli'])
```

```python
# plotting multiple plots

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])
```

```python
# labels title

plt.plot(batsman['index'], batsman['V Kohli'])
plt.plot(batsman['index'], batsman['RG Sharma'])

plt.title('Rohit Sharma Vs Virat Kohli IPL runs comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# colors (hex) and line (width and style) and marker (size)

plt.plot(batsman['index'], batsman['V Kohli'], color='green')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#D9F123')

plt.title('Rohit Sharma Vs Virat Kohli IPL runs comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', linewidth=3)
plt.plot(batsman['index'], batsman['RG Sharma'], color='#888', linestyle='dashdot', linewidth=1)

plt.title('Rohit Sharma Vs Virat Kohli IPL runs comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', linewidth=3, marker='d', 
         markersize=15)
plt.plot(batsman['index'], batsman['RG Sharma'], color='#888', linestyle='dashdot', linewidth=1, marker='+')

plt.title('Rohit Sharma Vs Virat Kohli IPL runs comparison')
plt.xlabel('Season')
plt.ylabel('Runs')
```

```python
# legend 

plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', linewidth=3, marker='d', 
         markersize=15, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#888', linestyle='dashdot', linewidth=1, marker='+',
        label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli IPL runs comparison')
plt.xlabel('Season')
plt.ylabel('Runs')

plt.legend(loc='upper right')
```

```python
# limiting axes

price = [45000, 54000, 34000, 49000, 60000, 55000, 4500000]
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021]

plt.plot(year, price)
plt.ylim(0, 100000)
plt.xlim(2017, 2019)
```

```python
# grid

plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', linewidth=3, marker='d', 
         markersize=15, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#888', linestyle='dashdot', linewidth=1, marker='+',
        label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli IPL runs comparison')
plt.xlabel('Season')
plt.ylabel('Runs')

plt.legend(loc='upper right')

plt.grid()
```

```python
# show

# kisi kisi text edior ya flask jaise module me graph show nhi hota hai toh plt.show() use karte hai.

plt.plot(batsman['index'], batsman['V Kohli'], color='green', linestyle='dashed', linewidth=3, marker='d', 
         markersize=15, label='Virat')
plt.plot(batsman['index'], batsman['RG Sharma'], color='#888', linestyle='dashdot', linewidth=1, marker='+',
        label='Rohit')

plt.title('Rohit Sharma Vs Virat Kohli IPL runs comparison')
plt.xlabel('Season')
plt.ylabel('Runs')

plt.legend(loc='upper right')

plt.grid()

plt.show()
```

## Plot-2: Scatter Plot

image.png

- Bivariate Analysis
- numerical vs numerical
- Use case - Finding correlation

```python
# plt.scatter simple function

x = np.linspace(-10,10,50)
y = 10*x + 3 + np.random.randint(0,300,50)
```

```python
plt.scatter(x, y)
```

```python
# plt.scatter on pandas data

batter = pd.read_csv(r'datasets/batter.csv')

batter.head(50)
```

```python
plt.scatter(batter['avg'], batter['strike_rate'], marker='+')

plt.title('Average vs Strike rate of top 50 IPL batsman')

plt.xlabel('average')
plt.ylabel('strike rate')
```

```python
# size

tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'], tips['tip'], color='green', s=tips['size']*20)

# s represents ki kitne logo ke sath aaya.
```

```python
# scatterplot using plt.plot

# plt.plot(tips['total_bill'], tips['tip']) # a 2d plot

plt.plot(tips['total_bill'], tips['tip'], 'o', color='red', marker='+') # scatter plot

# other features s cannot be used.

# But its a faster technique than plt.scatter().

# For large datasets use plot
# For smaller datasets use scatter
```

## PLot-3: Bar Chart

image.png

- x axis - categorrical data
- y axis - aggregate (e.g. count, avg age)

- Bivariate Analysis
- Numerical vs Categorical
- Use case - Aggregate analysis of groups

### simple bar chart

```python
children = [10, 20, 15, 30, 25]
colors = ['red', 'blue', 'green', 'pink', 'yellow']

plt.bar(colors, children)
```

### horizontal bar chart

```python
plt.barh(colors, children)
```

```python
# if number of categories more use horizontal
# if number of categories less use vertical
```

### multiple bar charts

```python
batsman = pd.read_csv(r'datasets/batsman_season_record.csv')
batsman
```

```python
plt.bar(batsman['batsman'], batsman['2015'])
plt.bar(batsman['batsman'], batsman['2016'])
plt.bar(batsman['batsman'], batsman['2017'])
```

```python
plt.bar(np.arange(batsman.shape[0]) - 0.2, batsman['2015'], width=0.2)
plt.bar(np.arange(batsman.shape[0]), batsman['2016'], width=0.2)
plt.bar(np.arange(batsman.shape[0]) + 0.2, batsman['2017'], width=0.2)

plt.xticks(np.arange(batsman.shape[0]), batsman['batsman'])

plt.show()
```

### a problem

```python
children = [10, 20, 15, 30, 25]
colors = ['red red red red', 'blue blue blue blue', 'green green green green',
          'pink pink pink pink', 'yellow yellow yellow yellow']

plt.bar(colors, children)
plt.xticks(rotation=50)
plt.show()
```

### stacked bar chart

```python
batsman
```

```python
plt.bar(batsman['batsman'], batsman['2015'], label='2015')
plt.bar(batsman['batsman'], batsman['2016'], bottom=batsman['2015'], label='2016')
plt.bar(batsman['batsman'], batsman['2017'], bottom=batsman['2015']+batsman['2016'], label='2017')

plt.legend()
plt.show()
```

## Pie Chart

image.png

- Univariate/Bivariate Analysis
- Categorical vs numerical
- Use case - To find contibution on a standard scale

### simple data

```python
data_ = [12, 34, 56, 32, 34, 56]

plt.pie(data_)
plt.show()
```

```python
marks = [95, 57, 85, 65, 45]
subjects = ['maths', 'physics', 'english', 'phy edu', 'chemistry']

plt.pie(marks, labels=subjects)
plt.show()
```

### dataset

```python
gayle = pd.read_csv(r'datasets/gayle-175.csv')
gayle
```

```python
plt.pie(gayle['batsman_runs'], labels=gayle['batsman'], autopct='%0.1f%%')
plt.show()
```

### colors

```python
plt.pie(gayle['batsman_runs'], labels=gayle['batsman'], autopct='%0.1f%%', 
        colors=['blue','green','pink','cyan','gray','red'])
plt.show()
```

### explode shadow

```python
plt.pie(gayle['batsman_runs'], labels=gayle['batsman'], autopct='%0.1f%%', explode=[0.2,0,0,0,0,0.1], shadow=True)
plt.show()
```
