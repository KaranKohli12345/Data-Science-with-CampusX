```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Colored Scatterplots

```python
iris = pd.read_csv(r'Datasets/iris.csv')
iris.head()
```

```python
iris.sample(10)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'])
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.show()
```

```python
iris['Species'] = iris['Species'].replace({
    'Iris-setosa': 0,
    'Iris-virginica': 1,
    'Iris-versicolor': 2,
})
```

```python
iris.sample(10)
```

```python
plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='winter', alpha=0.5)
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.colorbar()
plt.show()
```

## Plot size

```python
plt.figure(figsize=(10,5))

plt.scatter(iris['SepalLengthCm'], iris['PetalLengthCm'], c=iris['Species'], cmap='winter', alpha=0.5)
plt.xlabel('Sepal length')
plt.ylabel('Petal length')
plt.colorbar()
plt.show()
```
