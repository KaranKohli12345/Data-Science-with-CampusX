### Example - ATM machine ka code

````python
# we write all class names by Pascal case
# e.g. HelloWorld 
````

````python
class ATM:

    # constructor (special method) --> superpower --> ise call karne ki jarurat nahi, 
    # khud hi execute ho jata hai, jaise hi class ka ek object banaya jata hai.
    def __init__(self):
        self.pin = ''
        self.balance = 0
        # print('Mai to bina call kare hi execute ho gaya!')
        self.menu()

    def menu(self):
        user_input = input('''
        Hi! how can I help You?
        1. Press 1 to create PIN
        2. Press 2 to change PIN
        3. Press 3 to check balance
        4. Press 4 to withdraw money
        5. Press any other number to exit
        ''')

        if user_input == '1':
            self.create_pin()
        elif user_input == '2':
            self.change_pin()
        elif user_input == '3':
            self.check_balance()
        elif user_input == '4':
            self.withdraw_money()
        else:
            exit()

    def create_pin(self):
        user_pin = input('Enter your pin: ')
        self.pin = user_pin

        user_balance = int(input('Enter balance: '))
        self.balance = user_balance

        print('PIN created successfully')
        self.menu()

    def change_pin(self):
        old_pin = input('Enter old PIN: ')
        
        if old_pin == self.pin:
            new_pin = input('Enter new PIN: ')
            self.pin = new_pin
            print('PIN changes succesfully')
        else:
            print('Nahi karne dunga.')
            
        self.menu()    

    def check_balance(self):
        user_pin = input('Enter your PIN: ')
        if self.pin == user_pin:
            print('Your balance is: ', self.balance)
        else:
            print('Chal nikal yaha se')
        self.menu()

    def withdraw_money(self):
        user_pin = input('Enter you PIN: ')
        if self.pin == user_pin:
            amount = int(input('Enter amount: '))
            if amount <= self.balance:
                self.balance -= amount
                print('Withdrawl successfully')
                print('Your balance is: ', self.balance)
            else:
                print('bhai paisa dalwa le')
        else:
            print('nahi dunga paisa')
        self.menu()
````

````python
obj = ATM()
````

````python
print(type(obj))
````

## Let's make our own custom data type called Fraction

````python
3/5*1/3
````

````python
class Fraction:

    # parameterized constructor
    # Magic method 1
    def __init__(self, x, y):
        self.num = x
        self.den = y

    # Magic method 2
    def __str__(self): # print ke andar object ko dalenge toh iske andar ka code execute ho jayega
        return '{}/{}'.format(self.num, self.den)

    # Magic method 3
    def __add__(self, other): # jab bhi object1 + object2 dikhe ye automatically execute ho jayega
        new_num = self.num * other.den + other.num * self.den
        new_den = self.den * other.den
        
        return '{}/{}'.format(new_num, new_den)

    # Magic method 4
    def __sub__(self, other): # jab bhi object1 - object2 dikhe ye automatically execute ho jayega
        new_num = self.num * other.den - other.num * self.den
        new_den = self.den * other.den
        
        return '{}/{}'.format(new_num, new_den)

    # Magic method 5
    def __mul__(self, other): # jab bhi object1 * object2 dikhe ye automatically execute ho jayega
        new_num = self.num * other.num
        new_den = self.den * other.den
        
        return '{}/{}'.format(new_num, new_den)

    # Magic method 6
    def __truediv__(self, other): # jab bhi object1 / object2 dikhe ye automatically execute ho jayega
        new_num = self.num * other.den
        new_den = self.den * other.num
        
        return '{}/{}'.format(new_num, new_den)

    # Non-magic method
    def convert_to_decimal(self):
        return self.num / self.den
        
````

````python
fr1 = Fraction()
````

````python
fr1 = Fraction(3,5)
fr2 = Fraction(1,2)
````

````python
print(type(fr1))
print(type(fr2))
````

````python
print(fr1)
print(fr2)
````

````python
print(fr1 + fr2)
print(fr1 - fr2)
print(fr1 * fr2)
print(fr1 / fr2)
print(fr1.convert_to_decimal())
print(fr2.convert_to_decimal())
````
