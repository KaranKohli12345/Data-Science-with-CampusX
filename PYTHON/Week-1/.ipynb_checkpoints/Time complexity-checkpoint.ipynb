## Ways to measure the time of execution of a program 

## 1. Measuring time simply

```python
import time

start = time.time() # time() tells current time at that moment.

for i in range(1,1001):
    print(i)

end = time.time() 

print(end - start)
```

```python
import time

start = time.time() # time() tells current time at that moment.

i = 1

while i <= 1000:
    print(i)
    i += 1

end = time.time() 

print(end - start)
```

```python
import time

start = time.time() # time() tells current time at that moment.

a = 2
b = 5

print(a + b)

end = time.time() 

print(end - start)
```

## input - 10 100 1000 ...
## time - 1 2 3 ...

## v). nlogn - sorting algorithms

## e.g. merge sort
