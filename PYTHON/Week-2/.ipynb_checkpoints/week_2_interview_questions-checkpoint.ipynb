### 2. What is garbage collection?

#### in python the values/literals that are not referenced/used by any variable waste memory because they are not used by any program, which are removed periodically by an internal program called garbage collector.

#### python doesnt give this freedom to us to change things at memory/hardware level so garbage collection is automatic.

### 3. What is mutability and why is it dangerous in certain scenarios?

```
Mutable    |  Immutable
----------------------------------------------------------
list       |  int, float, string, boolean, tuple, complex
set        |
dictionary |
```

- The ability to change a data type's value by residing on its memory location is called mutability.
- It depends on datatype.

````python
# mutable

L = [1,2,3]

print(id(L))

L.append(4)
print(L)
print(id(L))
````

````python
# immutable

T = (1,2,3)

print(id(T))

T = T + (4,)

print(T)
print(id(T))
````

#### why is it dangerous in certain scenarios?

````python
a = [1,2,3,4]
b = a

b.append(5)
print(b)

print(a)
````

````python
def func(data):
    data.append(4)

a = [1,2,3]
func(a)
print(a)
````

````python
def func(data):
    data += (4,5)

a = (1,2,3)
func(a)
print(a)
````

### 4. What is cloning?

````python
a = [1,2,3]

# cloning
b = a[:]
````

````python
print(id(a))
print(id(b))
````

````python
b.append(4)
print(b)
print(a)
````

#### how to avoid above problem?

````python
def func(data):
    data.append(4)

a = [1,2,3]
func(a[:])
print(a)
````

````python
# or

def func(data):
    data.append(4)

a = [1,2,3]
func(a.copy())
print(a)
````

````python
a = {'name': 'python', 'age': 30}
b = a.copy()

b['gender'] = 'male'

b
````

````python
a
````

### 5. Differentiate between deep and shallow copies.

#### shallow copy

````python
a = [1,2,3]
# b = a[:]
# or
b = a.copy() # shallow copy

b.append(4)

print(b)
print(a)
````

````python
a = [1,2,3,[4,5]]
a
````

````python
b = a.copy()
b
````

````python
print(id(a))
print(id(b))
````

````python
b[-1][0] = 400
b
````

````python
a
````

````python
print(id(a[-1]))
print(id(b[-1]))
````

#### deep copy

````python
import copy

a = [1,2,3,[4,5]]

b = copy.deepcopy(a)

b
````

````python
b[-1][0] = 400
b
````

````python
a
````

````python
print(id(a[-1]))
print(id(b[-1]))
````
